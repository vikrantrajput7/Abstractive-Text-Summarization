# Creating Vocabulary

Max Input sequence length is 110 and max output sequence length is 26
We are taking top 230000 frequent words in Source Vocabulary and top 85000 frequent words in target vocabulary

```python
!python preprocess.py -train_src "/content/OpenNMT-py/data/drive/My Drive/Final Vikrant Project /t-src-train.txt" -train_tgt "/content/OpenNMT-py/data/drive/My Drive/Final Vikrant Project /t-tgt-train.txt" -valid_src "/content/OpenNMT-py/data/drive/My Drive/Final Vikrant Project /t-src-val.txt" -valid_tgt "/content/OpenNMT-py/data/drive/My Drive/Final Vikrant Project /t-tgt-val.txt" -save_data data/data --src_seq_length 110 --src_seq_length_trunc 110 --tgt_seq_length 26 --tgt_seq_length_trunc 26 --src_vocab_size 230000 --tgt_vocab_size 85000  -overwrite
```

# Training the model

* Encoder is a single layer bi-Directional LSTM with ouput of 512 dimensions
* Decoder is a uni-Directional LSTM with ouput of 512 dimensions
* Validation Accuracy will be calculated on every 1000 epochs
* Model will be saved on every 1000 epochs
* Dropout rate is 0.5
* Batch Size is 128
* Optimizer : AdaGrad
* Model is trained for total of 90000 epochs
* Highest Validation accuracy is 37.1663 on 89000 epoch

```python
!python train.py -data data/data -save_model "/content/OpenNMT-py/data/drive/My Drive/Final Vikrant Project /demo-model" \
           --valid_steps 1000 --valid_batch_size 128 --save_checkpoint_steps 1000 \
           -global_attention mlp \
           -word_vec_size 128 \
           -rnn_size 512 \
           -layers 1 \
           -encoder_type brnn \
           -train_steps 70000 \
           -max_grad_norm 2 \
           -dropout 0.4 \
           -batch_size 128 \
           -optim adagrad \
           -learning_rate 0.15 \
           -adagrad_accumulator_init 0.1 \
           -gpu_ranks 0 
           #-train_from '/content/OpenNMT-py/data/drive/My Drive/demo-model_step_37000.pt'
```

Continuing model training after 70000 epochs to 90000 epochs

```python
!python train.py -data data/data -save_model "/content/OpenNMT-py/data/drive/My Drive/Final Vikrant Project /demo-model-mlp" \
           --valid_steps 1000 --valid_batch_size 128 --save_checkpoint_steps 1000 \
           -global_attention mlp \
           -word_vec_size 128 \
           -rnn_size 512 \
           -layers 1 \
           -encoder_type brnn \
           -train_steps 90000 \
           -max_grad_norm 2 \
           -dropout 0.4 \
           -batch_size 128 \
           -optim adagrad \
           -learning_rate 0.15 \
           -adagrad_accumulator_init 0.1 \
           -gpu_ranks 0 \
           -train_from "/content/OpenNMT-py/data/drive/My Drive/Final Vikrant Project /demo-model_step_70000.pt"
```
